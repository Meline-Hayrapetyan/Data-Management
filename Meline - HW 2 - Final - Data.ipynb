```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

# *Scraping Information for one quote* 

```python
url = "http://quotes.toscrape.com"

response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")
```

```python
quote = soup.find("div", class_="quote")

quote_text = quote.find("span", class_="text").get_text()
author = quote.find("small", class_="author").get_text()
tags = [tag.get_text() for tag in quote.find_all("a", class_="tag")]


print("Quote:", quote_text)
print("Author:", author)
print("Tags:", tags)
```

```python
author_link = quote.find("a")["href"]


author_page = requests.get(url + author_link)
author_soup = BeautifulSoup(author_page.content, "html.parser")
```

```python
born = author_soup.find("span", class_="author-born-date").get_text()
born_location = author_soup.find("span", class_="author-born-location").get_text()
description = author_soup.find("div", class_="author-description").get_text(strip=True)


print("Born:", born)
print("Born Location:", born_location)
print("Description:", description)
```

# *Scraping all quotes*

```python
base_url = "http://quotes.toscrape.com"

all_quotes = []

page_url = "/page/1/"
while page_url:
    
    response = requests.get(base_url + page_url)
    soup = BeautifulSoup(response.content, "html.parser")
    
    quotes = soup.find_all("div", class_="quote")
    
    for quote in quotes:
        quote_text = quote.find("span", class_="text").get_text()
        author = quote.find("small", class_="author").get_text()
        tags = [tag.get_text() for tag in quote.find_all("a", class_="tag")]
        
        author_link = quote.find("a")["href"]
        author_page = requests.get(base_url + author_link)
        author_soup = BeautifulSoup(author_page.content, "html.parser")
        
        born = author_soup.find("span", class_="author-born-date").get_text()
        born_location = author_soup.find("span", class_="author-born-location").get_text()
        description = author_soup.find("div", class_="author-description").get_text(strip=True)
        

        all_quotes.append({
            "quote": quote_text,
            "author": author,
            "tags": ", ".join(tags),
            "born": born,
            "born_location": born_location,
            "description": description
        })
    
   
    next_page = soup.find("li", class_="next")
    page_url = next_page.find("a")["href"] if next_page else None

```

```python
df = pd.DataFrame(all_quotes)
df.to_csv("quotes_scraped.csv", index=False)

print(df.head())
```
